```python
def fetch_medal_tally(df, year, country):
    medal_df = df.drop_duplicates(subset= ['Team', 'NOC', 'Games', 'Year', 'City', 'Sport', 'Event' , 'Medal'])
    flag = 0
    if year == 'Overall' and country == 'Overall':
        temp_df = medal_df

    elif year == 'Overall' and country != 'Overall':
        flag = 1
        temp_df = medal_df[medal_df['region'] == country]

    elif year != 'Overall' and country == 'Overall':
        temp_df = medal_df[medal_df['Year'] == int(year)]

    else:  # year != 'Overall' and country != 'Overall'
        flag = 2
        temp_df = medal_df[(medal_df['Year'] == int(year)) & (medal_df['region'] == country)]

    # group depending on flag
    if flag == 1:  # single country → show progress over years
        x = (
            temp_df.groupby('Year')
            .sum()[['Gold', 'Silver', 'Bronze']]
            .sort_values('Year')
            .reset_index()
        )
    else:  # overall or single year → show by region
        x = (
            temp_df.groupby('region')
            .sum()[['Gold', 'Silver', 'Bronze']]
            .sort_values('Gold', ascending=False)
            .reset_index()
        )

    # add total column in both cases
    x['total'] = x['Gold'] + x['Silver'] + x['Bronze']

    return x
```

```python
fetch_medal_tally(df,year='Overall',country= 'India')
```
